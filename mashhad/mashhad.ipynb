```python
import os

import numpy as np
import pandas as pd

from shapely import wkt
from shapely.geometry import Polygon
import requests

def polygon_to_list(polygon):
    return list(polygon.exterior.coords)


category = [{'public transport': ['bureau_de_change', 'bus_station', 'car_sharing', 'busway', 'road', 'track',
                                  'motorway', 'path', 'motorcycle_parking', 'parking', 'car_rental', 'corridor',
                                  'taxi', 'fuel', 'parking_entrance']},
            {'cycling infrastructure': ['bicycle_parking', 'bicycle_rental', 'cycleway']},
            {'sidewalks': ['footway', 'pedestrian']},
            {'residential': ['residential']},
            {'public safety': ['fire_station', 'police']},
            {'commercial': ['atm', 'bank', 'register_office', 'post_box', 'post_office', 'marketplace',
                            'vending_machine']},
            {'educational': ['college', 'university', 'driving_school', 'school', 'kindergarten', 'music_school']},
            {'recreational': ['cafe', 'cinema', 'fast_food', 'restaurant', 'ice_cream', 'food_court', 'rest_area',
                              'drinking_water', 'internet_cafe', 'library', 'place_of_worship', 'social_centre',
                              'service', 'social_facility', 'theatre', 'toilets', 'townhall', 'waste_basket',
                              'waste_disposal']},
            {'healthcare': ['dentist', 'doctors', 'clinic', 'pharmacy', 'hospital', 'veterinary']}]

keywords = ['atm', 'bureau_de_change', 'bus_station', 'car_sharing', 'busway', 'road', 'track',
            'motorway', 'path', 'motorcycle_parking', 'parking', 'car_rental', 'corridor', 'vending_machine'
            'taxi', 'fuel', 'parking_entrance', 'bicycle_parking', 'bicycle_rental', 'music_school'
            'cycleway', 'footway', 'pedestrian', 'residential', 'fire_station', 'police', 'bank', 'register_office',
            'post_box', 'post_office', 'college', 'university', 'driving_school', 'school', 'kindergarten',
            'cafe', 'cinema', 'fast_food', 'restaurant', 'ice_cream', 'food_court', 'rest_area',
            'drinking_water', 'internet_cafe', 'library', 'place_of_worship', 'social_centre', 'service',
            'social_facility', 'theatre', 'toilets', 'townhall', 'waste_basket', 'waste_disposal', 'marketplace',
            'dentist', 'doctors', 'clinic', 'pharmacy', 'hospital', 'veterinary']

results = {}
count = []

for path, subdirs, files in os.walk('/content/drive/MyDrive/Colab Notebooks/amenity/'):
    for i in range(len(files)):
        try:
            # print('================== start')
            # print(files[i].split('-'))
            key = files[i].split('-')[0]
            relation_id = files[i].split('-')[1]
            name = files[i].split('-')[2]

            a = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/amenity/' + files[i])

            if name == 'AmentiesCount.csv':
                name = 'amenity'
            elif name == 'RoadLength.csv':
                name = 'highway'

            common_amenities = [{keywords[i]: a['count'].iloc[a[name].to_list().index(keywords[i])]}
                                for i in range(len(keywords))
                                if keywords[i] in a[name].to_list()]
            # print(common_amenities)

            count = []

            for i, cat in enumerate(category):
                cat_count = 0
                for amenity in common_amenities:
                    if list(amenity.keys())[0] in list(cat.values())[0]:
                        cat_count += list(amenity.values())[0]
                count.append(cat_count)

            item = results.get(key)
            if item is not None:
                count = [c + i for c, i in zip(count, item)]

            total_count = len(count)
            percentage = (sum(1 for c in count if c >= 1) / total_count) * 100
            count.append(round(percentage, 2))

            try:
                overpass_url = "https://overpass-api.de/api/interpreter"
                query = f"""
                [out:json];
                relation({relation_id});
                out geom;
                """

                response = requests.get(overpass_url, params={"data": query})
                res = response.json()
                print(res)
                relation = res["elements"][0]

                # geometry = relation['members'][0]['geometry']

                # geometry = relation['bounds']
                # count.append(geometry)

                # print(key, geometry)
                # print('================== end')

                coordinates = []
                for member in relation['members']:
                    if member['role'] == 'outer':
                      try:
                          coords = [(point['lat'], point['lon']) for point in member['geometry']]

                          polygon = Polygon(coords)
                          convex_hull = polygon.convex_hull
                          hull_coords = polygon_to_list(convex_hull)

                          coordinates.extend(hull_coords)

                      except:
                          pass

                count.append(coordinates)

            except: # ex:
                # print(ex)
                count.append([])

            results.__setitem__(key, count)

        except: #exp:
            # print(exp)
            print('error')

# categories = ['zone']
categories = [list(dic.keys())[0] for dic in category]
categories.append('percentage')
categories.append('coordinates')

df = pd.DataFrame(results, index=categories)
df = df.transpose()

total_results = len(results)
percentages = []
for key in categories:
    try:
        percentages.append(round((sum(1 for res in df[key] if res >= 1) / total_results) * 100, 2))
    except:
        percentages.append(0)

percentages.pop()
percentages.append(np.mean(df['percentage']))
results.__setitem__('percentage', percentages)
print(total_results)
df = pd.DataFrame(results, columns=results.keys(), index=categories)
# print(df.head())
df = df.transpose()
# df.to_html('results.html')
df.to_csv('/content/drive/MyDrive/Colab Notebooks/results-mashhad.csv', encoding='utf-8')
# df.to_excel('results.xlsx')
```

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import folium
from shapely.geometry import Polygon, MultiPolygon
from shapely.ops import unary_union
import geopandas as gpd

def polygon_to_list(polygon):
    return list(polygon.exterior.coords)

# دیتاست را بارگیری کنید (من فرض می‌کنم دیتاست در یک فایل CSV به نام 'data.csv' ذخیره شده است)
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/results-mashhad.csv')

# تبدیل رشته‌های مختصات به لیست
data['coordinates'] = data['coordinates'].apply(lambda x: ast.literal_eval(x))

# ایجاد نقشه با استفاده از folium
m = folium.Map(location=[36.3, 59.5], zoom_start=11)
# center_lat = sum([coord[0] for polygon in coordinates for coord in polygon]) / sum([len(polygon) for polygon in coordinates])
# center_lon = sum([coord[1] for polygon in coordinates for coord in polygon]) / sum([len(polygon) for polygon in coordinates])
# m = folium.Map(location=[center_lat, center_lon], zoom_start=12)


# تعیین رنگ بر اساس درصد و ایجاد چند ضلعی‌ها بر روی نقشه
for index, row in data.iterrows():
    try:
        percentage = row['percentage']
        coordinates = row['coordinates']
        # coordinates = [(row['coordinates']['minlat'], row['coordinates']['minlon']), (row['coordinates']['maxlat'], row['coordinates']['maxlon'])]
        #  [(coord['lat'], coord['lon']) for coord in row['coordinates']]

        polygon = Polygon(coordinates)
        convex_hull = polygon.convex_hull
        hull_coords = polygon_to_list(convex_hull)

        color = 'green' if percentage >= 80 else 'blue' if percentage >= 60 else 'yellow' if percentage >= 40 else 'red'
        folium.Polygon(locations=hull_coords, color=color, fill=True, fill_opacity=0.5).add_to(m)
    except:
        pass

# ذخیره نقشه به فایل HTML
m.save('/content/drive/MyDrive/Colab Notebooks/mashhad-map.html')

# Save the map as a PNG image directly
image_path = '/content/drive/MyDrive/Colab Notebooks/mashhad-map.png'
m.save(outfile=image_path)
```
